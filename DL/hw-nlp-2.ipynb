#Код для проверки

import re

n = int(input())

pat = re.compile(r'[^_]http[s]?:\/\/(?:www\.)?([\w\.-]+\.\w+)')

dom = set()

for i in range(n):
    inp = input()
    
    x = re.findall(pat,inp)
    x = set(x)
    for y in x:
        dom.add(y)
        
st = ''
for v,d in enumerate(sorted(list(dom))):
    if v+1 !=len(dom):
        st+=d +str(';')
    else:
        st+=d
print(st)

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
test = com.to_list()
```

```python
count = CountVectorizer(stop_words = stopword)

word_count = count.fit_transform(test)

print(word_count)
```

```python
print(word_count.toarray())
```

```python
tfidf_transformer = TfidfTransformer()

tfidf_transformer.fit(word_count)

df_idf = pd.DataFrame(tfidf_transformer.idf_,index = count.get_feature_names_out(), columns = ['idf'])
```

```python
df_idf.sort_values(by = ['idf'])
```

```python
tf_idf_vector = tfidf_transformer.transform(word_count)

feauter_names = count.get_feature_names_out()

first_name = tf_idf_vector[50]

df_tfidf = pd.DataFrame(first_name.T.todense(), index = feauter_names, columns=['tfidf'])


df_tfidf.sort_values(by = ['tfidf'], ascending=False).iloc[:20]
```

```python
data1 = list(open('./quora.txt', encoding='utf-8'))

data1 = data1[:100]
```

```python
tf_v_e = TfidfVectorizer()

tf_m = tf_v_e.fit_transform(data1)
```

```python
size = len(tf_m.toarray())

arr = tf_m.toarray().tolist()
```

```python
res = []
for i in range(size):
    for j in range(size):
        if arr[i] != arr[j]:
            cos = dot(arr[i], arr[j])/(norm(arr[i])*norm(arr[j]))
            res+=[[cos,data1[i],data1[j]]]
```

```python
sorted(res, key = lambda x: x[0],reverse=True)[:1]
```
