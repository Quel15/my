```python
import pandas as pd
import numpy as np
import re
```

```python
def max_s(df):
    """Нахождение максимального значения"""
    if df['types']=='в пределах' or (df['types']!='в пределах' and len(df['val']) > 1) :
        return df['val'][1]
    else:
        return df['val'][0]
    
    
def check_v(df):
    """Сравнение результатов анализа"""
    if df['Результат анализа'] > df['maxs']: return 'Выше нормы'
    elif df['Результат анализа'] < df['mins']: return 'Ниже нормы'
    else: return "Норма"
```

```python
y = pd.read_csv('Химический анализ родника в Нескучном саду.csv', sep = ';')
```

```python
y.loc[y['Показатель']=='Цветность','Результат анализа'] = 0

y.loc[y['Показатель']=='Цветность','Норматив'] = 'не более 0'

y['Норматив'] = y['Норматив'].map(lambda x: x.replace(',','.'))

y['val'] = y.Норматив.map(lambda x: re.findall(r'\d+\.*\d*', x))

y['types'] = y.Норматив.map(lambda x: re.findall(r'\w+\s\w+', x)[0])

y
```

```python
y = y.assign(mins = y.apply(lambda x: x['val'][0] if len(x.val) > 1 and x['types']=='в пределах' else 0, axis=1),
    maxs = y.apply(lambda x: max_s(x), axis=1))

y
```

```python
y = y.astype({'mins':float, 'maxs':float, 'Результат анализа': float})
```

```python
y['test'] = y.apply(lambda x: check_v(x), axis=1)
```

```python
y = y.set_index('Показатель')
```

```python
y.drop(columns=['Норматив','val'])
```

Задача 2
В ящике 5 апельсинов и 4 яблока. Наудачу выбираются 3 фрукта. Какова вероятность, что все три фрукта – апельсины?

В интернете полученный аналитически ответ 0.119. Подтверждается ли он эксперементально?

```python
box = ['A']*4 + ['O']*5

for i in [10, 100, 1000, 10000, 100000, 1000000]:
    fr = pd.DataFrame(list(
                        map(lambda x:
                           np.random.choice(box, 3, replace = False),
                           range(i))), columns=['first', 'second', 'third'])
    
    org = fr.query('first == "O" and second =="O" and third == "O"').shape[0]
    
    print(f'Вероятность вытащить 3 апельсина при {i} попытках {org/fr.shape[0]}')
```

Задача 3
Мастер, имея 10 деталей, из которых 3 – нестандартных, проверяет детали одну за другой, пока ему не попадется стандартная. Какова вероятность, что он проверит ровно две детали?

В интернете полученный аналитически ответ 7/30 или 0.23333. Подтверждается ли он эксперементально?

```python
boxp = ['S']*7 + ['N']*3


for i in [10, 100, 1000, 10000, 100000, 1000000]:
    d = pd.DataFrame(list(
                        map(lambda x:
                           np.random.choice(boxp, 2, replace = False),
                           range(i))), columns=['first', 'second'])
    
    itog = d.query('first == "N" and second =="S"').shape[0]
    
    print(f'Вероятность проверить 2 детали при {i} попытках {itog/d.shape[0]}')
```
